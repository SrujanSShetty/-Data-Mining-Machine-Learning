```python
#11). Repeat b), c), d), e) in Q2, by using Python on the HW1.csv data set.

#Import Python Libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
from IPython.display import display, HTML
```

```python
df=pd.read_csv('HW1.csv')
```

```python
df.head()
```

```python
#b)Use min-max normalization to transform the values of the Income attribute onto the range [1-5].

# Normalization method  Min-max normalization [1,5]
df["Income_Min-max"]=((df["Income"]-df["Income"].min())/(df["Income"].max()-df["Income"].min()))*(4)+1
display(HTML(df.head(50).to_html()))
```

```python
#c)Use z-score normalization to standardize the values of the age attribute.

from scipy.stats import zscore
df[['Age_z-score']] = zscore(df[['Age']])
display(HTML(df.head(50).to_html()))

```

```python
# Define the bins and labels
bins = [1, 21, 41, float('inf')]
labels = ['Young', 'MidAge', 'Old']
# Creating a new column called 'Age_Category' for discretized values
df['Age_discretization'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
display(HTML(df.head(50).to_html()))
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
df2=pd.read_csv('HW1.csv')
# Use LabelEncoder to convert the nominal variable 'Gender' to binary (0 for 'F', 1 for 'M')
le = LabelEncoder()
df2['Gender'] = le.fit_transform(df2['Gender'])
print(df2)
#We can also use if loop to get a similar output
```

```python
le = LabelEncoder()
df2['Incidentals'] = le.fit_transform(df2['Incidentals'])
print(df2)
```

```python


df = pd.DataFrame(df2)

# Initialize new columns with 0
df['Genre_Action'] = 0
df['Genre_Comedy'] = 0


for index, row in df.iterrows():
    if row['Genre'] == 'Action':
        df.at[index, 'Genre_Action'] = 1
    elif row['Genre'] == 'Comedy':
        df.at[index, 'Genre_Comedy'] = 1


df = df.drop('Genre', axis=1)
print(df)
```

```python
2)numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64'] 
# get column names 
cols_numeric = df.select_dtypes(include=numerics).columns.tolist() 
df_numeric = df[cols_numeric] 
display(HTML(df.head(5).to_html())) 
corr=df_numeric.corr(method='pearson') 
display(HTML(corr.to_html()))
```

```python
df=pd.read_csv('HW1.csv')
from scipy.stats import f_oneway  

genre_action = df[df['Genre']=='Action']['Age']
genre_drama = df[df['Genre']=='Drama']['Age']
genre_comedy = df[df['Genre']=='Comedy']['Age']  
print(f_oneway(genre_action, genre_drama,genre_comedy))
```

```python
from scipy.stats import f_oneway  

genre_action = df[df['Genre']=='Action']['Rentals']
genre_drama = df[df['Genre']=='Drama']['Rentals']
genre_comedy = df[df['Genre']=='Comedy']['Rentals']  
print(f_oneway(genre_action, genre_drama,genre_comedy))
```

```python
from scipy.stats import f_oneway  

genre_action = df[df['Genre']=='Action']['Income']
genre_drama = df[df['Genre']=='Drama']['Income']
genre_comedy = df[df['Genre']=='Comedy']['Income']  
print(f_oneway(genre_action, genre_drama,genre_comedy))
```
